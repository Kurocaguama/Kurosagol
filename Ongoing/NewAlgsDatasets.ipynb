```python
from datasets import load_dataset

ds = load_dataset("trl-lib/DeepMath-103K")
ds
```

```python
prompts = ds['train']['prompt']
solution = ds['train']['solution']

avg_p_len = 0
avg_sol_len = 0
for i in range(len(prompts)):
    avg_p_len+=len(prompts[i][0]['content'])
    avg_sol_len+=len(solution[i])

print("Longitud promedio (en caracteres) del prompt: {}".format(round(avg_p_len/len(prompts), 3)))
print("Longitud promedio (en caracteres) de la respuesta: {}".format(round(avg_sol_len/len(prompts), 3)))
# XD
```
