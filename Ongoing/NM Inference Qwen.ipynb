```python
import pandas as pd

# Obs: Estas generaciones son de Qwen.
inf_ds = pd.read_csv("/home/flopezp/Kurosagol/Ongoing/datasets/qwen_inference_results.csv")
folio = pd.read_json("/home/flopezp/Kurosagol/FOLIO/FOLIO/folio_test.jsonl", lines=True)
inf_ds.head()
```

```python
from sklearn.metrics import classification_report, confusion_matrix
qwen_list = []
folio_list = []
differences = []
for i in range(len(inf_ds["Inference"])):
    split = inf_ds["Inference"].iloc[i].split()
    norm_qwen = split[0].lower()
    norm_folio = folio["label"][i].lower()
    if norm_qwen == 'true':
        qwen_list.append(1)
    if norm_qwen == 'false':
        qwen_list.append(0)
    if norm_qwen == 'uncertain':
        qwen_list.append(2)
    if norm_folio == 'true':
        folio_list.append(1)
    if norm_folio == 'false':
        folio_list.append(0)
    if norm_folio == 'uncertain':
        folio_list.append(2)
    
    if norm_qwen != norm_folio:
        differences.append(i)

print(classification_report(qwen_list, folio_list))
print(confusion_matrix(qwen_list, folio_list))
print("Cantidad de diferencias: {}".format(len(differences)))
```
